```
import pandas as pd # matris ve veri okuma işlerinde kullanılır
import numpy as np
from sklearn.preprocessing import LabelEncoder # Etiketleri numaralandırma işleminde kullanılır
import tensorflow as tf 
from tensorflow import keras #model oluşturulma aşamasında kullanırılır
from tensorflow.keras import layers # yapay zeka gerçekleştririlirken katman oluşmasında kullanılır 
from sklearn.linear_model import LogisticRegression # logistik regrasyon sınıflandırılmasında kullanılır
import matplotlib.pyplot as plt # veriyi ve sonuçları görselleşmesinde yararlanılır
import math
import seaborn as sns
```

```
doh = pd.read_csv('doh.csv', encoding='iso-8859-1', sep=";") # doh veriseti okama işlemi gerçekleşir
benign = pd.read_csv('benign.csv', encoding='iso-8859-1', sep=";")# benign veriseti okama işlemi gerçekleşir
malicious = pd.read_csv('malicious.csv', encoding='iso-8859-1', sep=";") # malicious veriseti okama işlemi gerçekleşir
```

```
nondoh = pd.read_csv('l1-nondoh.csv', encoding='iso-8859-1', sep=",") # nondoh veriseti okama işlemi gerçekleşir
nondoh.head()
```

```
bir1 = pd.concat([benign,malicious]) # birinci model oluşumu için benign ve malicious verisetleri birleşir
colums = bir1.columns
bir1.describe(include = "all")
```

```
bir1.head()
```

```
dene = bir1
for i in range(0,33): # verisetindeki değişkenler numaralandıralar optimizasyon işlemi başlatılır
  dene[colums[i]] = dene[colums[i]].astype("category").cat.codes
dene.head()
```

```
dene.describe(include="all")
```

```
dene[colums[34]] = dene[colums[34]].astype("category").cat.codes
dene.head()
```

```
dene1 = dene # Sonuca etki etmeye veya olumsuz etki edebilecek veriler silinir
for i in range(5,34):
  dene1 = dene1.drop(colums[i], axis=1)
dene1.head()
```

```
dene1 = dene1.drop("TimeStamp", axis=1)
dene1.head()
```

```
X = dene1.iloc[:,:-1] # veri setinden ulaşmak istenen sonuç harici veriler X değişkenine atanır
Y = dene1.iloc[:,-1] # tespit edimek istenen durumlar Y değişkenine atanır
X.head()
```

```
import sklearn.model_selection as model_selection
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, Y, train_size=0.65,test_size=0.35, random_state=101) #veri seti model eğitimi ve testi olmak üzere %35 i test geri kalanı eğitimi için ayrılır
X_train.head()
```

```
logmodel = LogisticRegression() #model sınıfından nesne oluşturulur
logmodel.fit(X_train,y_train) #model sınıfına verileri girilerek modek öğrenimine başlanır
```

```
predictions = logmodel.predict(X_test) # model test edilerek kalitesi ölçülür
```

```
col = dene1.columns
col
```

```
from sklearn.metrics import classification_report
a = classification_report(y_test,predictions,target_names=['SourceIP', 'DestinationIP']) #model sonucunu raporlayarak değişkenler incelenir
```

```
X_train.head()
```

```
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,predictions) #model sonuçları doğruluk matrisinde gösterlerek değerlendirilir
```

```
from sklearn.metrics import accuracy_score # modelin genel sonucuna ulaşılır
accuracy_score(y_test,predictions)
```
